```python
# Title :- Implement Gradient Descent Algorithm to find the local minima of a function.
```

```python
cur_x = 2 
iters = 0 
max_iters = 10000 
learning_rate = 0.01 
precision = 0.000001 
previous_step_size = 1 
df = lambda x: 2*(x+3) 
```

```python
print(cur_x)
```

```python
x_history =[cur_x]
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - learning_rate * df(prev_x) #Grad descent
    x_history.append(cur_x)
    previous_step_size = abs(cur_x - prev_x) #Change in x
    iters = iters+1 #iteration count
    print("Iteration",iters,"\nX value is",cur_x) #Print iterations
    
print("The local minimum occurs at", cur_x)
```

```python
print("The local minimum occurs at", cur_x)
```

```python
# Create a range of x values to plot
x_vals = np.linspace(-10, 4, 100)
y = lambda x: x**2 + 6*x +9

# Plot the lamda function y
plt.plot(x_vals, y(x_vals))

# Plot the values of x at each iteration
plt.plot(x_history, y(np.array(x_history)), 'rx')

# Label the axes and add a title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')

# Show the plot
plt.show()
```
